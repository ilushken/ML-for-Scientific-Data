## Task 1.1 – Measure

We have classification problem. So the simplest measure is accuracy, but it will not work good if we have different numbers of cases for each type of apple. So, we check number of good and bad apples in the dataset.

```python
good_num = np.sum(data['Quality']=="good")
print("In dataset there are " + str(good_num) + \
      " good apples and " + str(data.shape[0] - good_num) + " bad apples.")
```

So the number of good and bad apples are pretty balanced, but just not to have a headache with balance in train/test datasets, we will use [F-score](https://en.wikipedia.org/wiki/F-score) measure defined as
$$
F_\beta = \dfrac{(1+\beta^2) \cdot \text{true positive}}{(1+\beta^2) \cdot \text{true positive} + \beta^2 \cdot \text{false negative} + \text{false positive}}
$$

We choose $\beta$ parameter based on the following considerations:
1) It is better to tell that apple is bad when it is good, that tell that it is good when it is bad because this will increase the number of satisfied clients
2) If some "good" apple is labeled as "bad" the consequences are not so bad (e.g. compared with cancer detection) because user will just take another apple in this case.
So recall is more important that presicion, and we take $\beta = 2$

We also can use [ROC-curve](https://en.wikipedia.org/wiki/Receiver_operating_characteristic), but $F_\beta$ seems simpler but sufficient to me for this task. So, in order to keep things simple we use $F_2$.

So, for both 2D and 3D we have overlapping... Therefore, no usefull information from PCA. Let us try [t-SNE](https://scikit-learn.org/stable/modules/generated/sklearn.manifold.TSNE.html)...

```python
perp_array = np.array([1, 3, 5, 10, 50, 100, 500, 1000])
features_tSNE = np.zeros([np.size(perp_array), np.shape(features_scaled)[0], 2])
for ind in range(np.size(perp_array)):
    model_tSNE = TSNE(n_components=2, init='random', learning_rate='auto', perplexity = perp_array[ind], random_state = 1024)
    features_tSNE[ind, :, :] = model_tSNE.fit_transform(features_scaled)
    print("Done for perplexity = " + str(perp_array[ind]))
```

```python
fig, axes = plt.subplots(nrows=2, ncols=4, figsize=(20, 8))
axes = axes.flatten()
for ind in range(np.size(perp_array)):
    tSNE_good = features_tSNE[ind, np.where(data["Quality"]==1), :]
    tSNE_bad = features_tSNE[ind, np.where(data["Quality"]==0), :]
    axes[ind].title.set_text("perplexity = " + str(perp_array[ind]))
    axes[ind].scatter(tSNE_good[0, :, 0], tSNE_good[0, :, 1], alpha=1)
    axes[ind].scatter(tSNE_bad[0, :, 0], tSNE_bad[0, :, 1], color='none', edgecolor='k', alpha=0.4)
    axes[ind].legend(["good", "bad"])

plt.plot()
```

### Task 1.3.1 – Simplest Baseline

For simplest baseline we will do following. In training dataset,
1) we choose $n$ features that have biggest correlation with quality.
2) We calculate the mean value for each of the feature separately for "good" and "bad" apples and use it as two $n$-dimensional vectors.
3) For each sample we construct similar vector, i.e. with coordinates having values of same features as we take for "mean" vectors.
4) We calculate distances (via $\ell_2$ norm) between sample and "good" vector and sample and "bad" vector.
5) The smallest distance will give us label, i.e. if for some sample x distance between sample and "good" vector is less that between sample and "bad" vector, then we say that this sample if "good".
